```python
import numpy as np
import matplotlib.pyplot as plt
import math

# 
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)

#  x 
x = np.arange(-7, 7, 0.01)

# 
params = [(0, 1), (0, 2), (3, 1)]

# 
y_values = [normal(x, mu, sigma) for mu, sigma in params]

# 
plt.figure(figsize=(8, 4))

# 
for y, (mu, sigma) in zip(y_values, params):
    plt.plot(x, y, label=f'mean {mu}, std {sigma}')

# 
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True)

#  Jupyter Notebook 
plt.show()
```

```python
#  Jupyter Notebook  matplotlib 
%matplotlib inline
import random
import torch
import matplotlib.pyplot as plt  #  matplotlib  d2l 

def synthetic_data(w, b, num_examples):
    """
    y = Xw + b + 
    :
        w (torch.Tensor):  (n,)n 
        b (float): 
        num_examples (int): 
    :
        X (torch.Tensor):  (num_examples, n)
        y (torch.Tensor):  (num_examples, 1)
    """
    #  X
    X = torch.normal(0, 1, (num_examples, len(w)))
    #  y = Xw + b
    y = torch.matmul(X, w) + b
    #  0 0.01 
    y += torch.normal(0, 0.01, y.shape)
    #  y 
    return X, y.reshape((-1, 1))

# 
true_w = torch.tensor([2, -3.4])
true_b = 4.2

# 
features, labels = synthetic_data(true_w, true_b, 1000)

# 
print('features:', features[0], '\nlabel:', labels[0])

# 
plt.figure(figsize=(8, 4))

# features  vs labels
plt.scatter(features[:, 1].detach().numpy(), labels.detach().numpy(), s=1)

# 
plt.xlabel('Feature 2')  # x 
plt.ylabel('Label')      # y 
plt.title('Synthetic Data: Feature 2 vs Label')  # 

# 
plt.show()
```

```python
def data_iter(batch_size, features, labels):
    """
    

    :
        batch_size (int): 
        features (torch.Tensor):  (num_examples, num_features)
        labels (torch.Tensor):  (num_examples, 1)

    :
        
    """
    # 
    num_examples = len(features)
    
    #  [0, 1, 2, ..., num_examples - 1]
    indices = list(range(num_examples))
    
    #  epoch 
    random.shuffle(indices)
    
    # 
    for i in range(0, num_examples, batch_size):
        # 
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)])
        
        # 
        yield features[batch_indices], labels[batch_indices]

# 
batch_size = 10

# 
for X, y in data_iter(batch_size, features, labels):
    # 
    print(X, '\n', y)
    break  # 
```

```python
def linreg(X, w, b): 
    """"""
    return torch.matmul(X, w) + b
```

## 

```python
def squared_loss(y_hat, y): 
    """"""
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

## 

```python
def sgd(params, lr, batch_size): #@save
    """"""
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

## 

```python
lr = 0.03
num_epochs = 3
net = linreg
loss = squared_loss
```

```python
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y) # Xy
        # l(batch_size,1)l
        # [w,b]
        l.sum().backward()
        sgd([w, b], lr, batch_size) # 
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

```python
def net(X):
    """
     Softmax 

    :
        X (torch.Tensor):  (batch_size, num_features)

    :
        torch.Tensor:  (batch_size, num_classes)
    """
    #  X 
    # X.reshape((-1, W.shape[0]))  X  (batch_size, num_features)
    # -1 
    X_reshaped = X.reshape((-1, W.shape[0]))

    #  X * W + b
    # torch.matmul(X_reshaped, W)  X  W 
    # + b  b
    linear_output = torch.matmul(X_reshaped, W) + b

    #  Softmax 
    # Softmax  1
    return softmax(linear_output)
```

```python
def cross_entropy(y_hat, y):
    """
    

    :
        y_hat (torch.Tensor):  (batch_size, num_classes)
        y (torch.Tensor):  (batch_size,)

    :
        torch.Tensor:  (batch_size,)
    """
    # 
    # y_hat[range(len(y_hat)), y]  y_hat 
    # range(len(y_hat))  0  batch_size-1 
    # y 
    selected_probs = y_hat[range(len(y_hat)), y]

    # -log(selected_probs)
    return - torch.log(selected_probs)
    
cross_entropy(y_hat, y)
```

```python
def accuracy(y_hat, y): #@save
    """"""
    #  y_hat 
    if len(y_hat.shape) > 1 and y_hat.shape[1] > 1:
        y_hat = y_hat.argmax(axis=1)
    
    # 
    cmp = y_hat.type(y.dtype) == y
    
    # 
    return float(cmp.type(y.dtype).sum())
```

```python
def evaluate_accuracy(net, data_iter): #@save
    """"""
    #  net  PyTorch  nn.Module 
    if isinstance(net, torch.nn.Module):
        net.eval()
    
    # 
    metric = Accumulator(2)
    
    # 
    with torch.no_grad():
        for X, y in data_iter:
            #  metric 
            metric.add(accuracy(net(X), y), y.numel())
    
    # 
    return metric[0] / metric[1]
```

Accumulatorevaluate_accuracyAccumulator2

```python
class Accumulator: #@save
    """n"""
    def __init__(self, n):
        #  n  0.0
        self.data = [0.0] * n
    
    def add(self, *args):
        #  self.data 
        self.data = [a + float(b) for a, b in zip(self.data, args)]
    
    def reset(self):
        #  0.0
        self.data = [0.0] * len(self.data)
    
    def __getitem__(self, idx):
        # 
        return self.data[idx]
```

```python
def train_epoch_ch3(net, train_iter, loss, updater): #@save
    # 
    if isinstance(net, torch.nn.Module):
        net.train()
    # 
    metric = Accumulator(3)
    for X, y in train_iter:
        # 
        y_hat = net(X)
        l = loss(y_hat, y)
        if isinstance(updater, torch.optim.Optimizer):
            # PyTorch
            updater.zero_grad()
            l.mean().backward()
            updater.step()
        else:
            # 
            l.sum().backward()
            updater(X.shape[0])
        metric.add(float(l.sum()), accuracy(y_hat, y), y.numel())
    # 
    return metric[0] / metric[2], metric[1] / metric[2]
```

```python
import matplotlib.pyplot as plt
from IPython import display

class Animator:
    """"""
    def __init__(self, xlabel=None, ylabel=None, legend=None, xlim=None,
                 ylim=None, xscale='linear', yscale='linear',
                 fmts=('-', 'm--', 'g-.', 'r:'), nrows=1, ncols=1,
                 figsize=(3.5, 2.5)):
        """
         Animator 

        :
            xlabel (str): x 
            ylabel (str): y 
            legend (list): 
            xlim (tuple): x 
            ylim (tuple): y 
            xscale (str): x  'linear'  'log'
            yscale (str): y  'linear'  'log'
            fmts (list):  '-' 'm--' 
            nrows (int): 
            ncols (int): 
            figsize (tuple): 
        """
        # 
        if legend is None:
            legend = []
        
        # 
        self.fig, self.axes = plt.subplots(nrows, ncols, figsize=figsize)
        if nrows * ncols == 1:
            self.axes = [self.axes, ]
        
        # 
        self.config_axes = lambda: self._set_axes(
            self.axes[0], xlabel, ylabel, xlim, ylim, xscale, yscale, legend)
        
        # 
        self.X, self.Y, self.fmts = None, None, fmts
    
    def _set_axes(self, ax, xlabel, ylabel, xlim, ylim, xscale, yscale, legend):
        """"""
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_xscale(xscale)
        ax.set_yscale(yscale)
        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
        if legend:
            ax.legend(legend)
        ax.grid()
    
    def add(self, x, y):
        """
        

        :
            x (float or list): x 
            y (float or list): y 
        """
        # 
        if not hasattr(y, "__len__"):
            y = [y]
        n = len(y)
        if not hasattr(x, "__len__"):
            x = [x] * n
        
        # 
        if not self.X:
            self.X = [[] for _ in range(n)]
        if not self.Y:
            self.Y = [[] for _ in range(n)]
        
        # 
        for i, (a, b) in enumerate(zip(x, y)):
            if a is not None and b is not None:
                self.X[i].append(a)
                self.Y[i].append(b)
        
        # 
        self.axes[0].cla()
        for x, y, fmt in zip(self.X, self.Y, self.fmts):
            self.axes[0].plot(x, y, fmt)
        
        # 
        self.config_axes()
        
        # 
        display.display(self.fig)
        display.clear_output(wait=True)
```

```python
def train_ch3(net, train_iter, test_iter, loss, num_epochs, updater): #@save
    """3"""
    #  Animator
    animator = Animator(xlabel='epoch', xlim=[1, num_epochs], ylim=[0.3, 0.9],
                        legend=['train loss', 'train acc', 'test acc'])
    
    # 
    for epoch in range(num_epochs):
        #  epoch
        train_metrics = train_epoch_ch3(net, train_iter, loss, updater)
        
        # 
        test_acc = evaluate_accuracy(net, test_iter)
        
        # 
        animator.add(epoch + 1, train_metrics + (test_acc,))
        
        # 
        train_loss, train_acc = train_metrics
        
        # 
        if epoch < num_epochs // 2:  # 
            assert train_loss < 1.0, f"Epoch {epoch + 1}: train_loss {train_loss} is too high"
        else:  # 
            assert train_loss < 0.5, f"Epoch {epoch + 1}: train_loss {train_loss} is too high"
        
        # 
        assert train_acc <= 1 and train_acc > 0.7, f"Epoch {epoch + 1}: train_acc {train_acc} is out of range"
        assert test_acc <= 1 and test_acc > 0.7, f"Epoch {epoch + 1}: test_acc {test_acc} is out of range"
```
