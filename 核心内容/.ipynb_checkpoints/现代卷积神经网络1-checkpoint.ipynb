# 深度卷积神经网络（AlexNet）


## 1. **AlexNet 的背景**
- **提出时间**: 2012 年
- **提出者**: Alex Krizhevsky、Ilya Sutskever 和 Geoffrey Hinton
- **比赛成绩**: 在 ImageNet 分类任务中，将 top-5 错误率从 26% 降低到了 15.3%，远超传统方法。
- **意义**: AlexNet 的成功证明了深度学习的潜力，开启了深度学习在计算机视觉领域的新时代。

---

## 2. **AlexNet 的网络结构**
AlexNet 是一个 8 层的神经网络，包括 5 个卷积层和 3 个全连接层。以下是其详细结构：

### 输入层
- 输入尺寸: $224 \times 224 \times 3$（实际输入为 $227 \times 227 \times 3$，原始论文中有误）
- 输入为 RGB 图像，经过预处理（如归一化）。

### 卷积层
1. **Conv1**:
   - 卷积核: $11 \times 11$，步幅 4，96 个滤波器
   - 输出尺寸: $55 \times 55 \times 96$
   - 激活函数: ReLU
   - 池化: $3 \times 3$ 最大池化，步幅 2
   - 输出尺寸: $27 \times 27 \times 96$

2. **Conv2**:
   - 卷积核: $5 \times 5$，填充 2，256 个滤波器
   - 输出尺寸: $27 \times 27 \times 256$
   - 激活函数: ReLU
   - 池化: $3 \times 3$ 最大池化，步幅 2
   - 输出尺寸: $13 \times 13 \times 256$

3. **Conv3**:
   - 卷积核: $3 \times 3$，填充 1，384 个滤波器
   - 输出尺寸: $13 \times 13 \times 384$
   - 激活函数: ReLU

4. **Conv4**:
   - 卷积核: $3 \times 3$，填充 1，384 个滤波器
   - 输出尺寸: $13 \times 13 \times 384$
   - 激活函数: ReLU

5. **Conv5**:
   - 卷积核: $3 \times 3$，填充 1，256 个滤波器
   - 输出尺寸: $13 \times 13 \times 256$
   - 激活函数: ReLU
   - 池化: $3 \times 3$ 最大池化，步幅 2
   - 输出尺寸: $6 \times 6 \times 256$

### 全连接层
6. **FC6**:
   - 输入: 将 $6 \times 6 \times 256$ 展平为 9216 维向量
   - 输出: 4096 维
   - 激活函数: ReLU
   - Dropout: 0.5

7. **FC7**:
   - 输入: 4096 维
   - 输出: 4096 维
   - 激活函数: ReLU
   - Dropout: 0.5

8. **FC8**:
   - 输入: 4096 维
   - 输出: 1000 维（对应 ImageNet 的 1000 个类别）
   - 激活函数: Softmax

---

## 3. **AlexNet 的主要特点**
1. **使用 ReLU 激活函数**:
   - 替代传统的 Sigmoid 或 Tanh 激活函数，ReLU 计算更快且缓解了梯度消失问题。

2. **使用 Dropout**:
   - 在全连接层中引入 Dropout，随机丢弃部分神经元，防止过拟合。

3. **数据增强**:
   - 通过对训练数据进行随机裁剪、翻转等操作，增加数据多样性，提升模型泛化能力。

4. **局部响应归一化（LRN）**:
   - 在卷积层后引入 LRN，增强局部特征的对比度（后续研究发现 LRN 效果有限，现代网络通常不使用）。

5. **多 GPU 训练**:
   - 由于网络规模较大，AlexNet 使用了两块 GPU 进行并行训练，提高了训练效率。

6. **重叠池化**:
   - 池化层的步幅小于池化核大小，减少了信息丢失。

---

## 4. **AlexNet 的贡献**
1. **证明了深度学习的潜力**:
   - AlexNet 的成功使深度学习成为计算机视觉领域的主流方法。

2. **推动了硬件发展**:
   - AlexNet 的训练需要大量计算资源，促进了 GPU 在深度学习中的应用。

3. **启发了后续网络设计**:
   - AlexNet 的设计思想（如 ReLU、Dropout、数据增强等）被广泛应用于后续的深度学习模型。

---

## 5. **AlexNet 的局限性**
1. **参数量大**:
   - 全连接层参数量巨大，导致模型存储和计算成本高。

2. **过拟合风险**:
   - 尽管使用了 Dropout 和数据增强，AlexNet 仍然容易过拟合，尤其是在小数据集上。

3. **LRN 效果有限**:
   - 局部响应归一化（LRN）的效果不如后续的 Batch Normalization。

---

## 6. **AlexNet 的现代意义**
- **经典模型**: AlexNet 是深度学习历史上的里程碑，至今仍被广泛学习和研究。
- **设计思想**: 其设计思想（如 ReLU、Dropout、数据增强等）被后续模型（如 VGG、ResNet 等）继承和发展。
- **教育价值**: AlexNet 是学习卷积神经网络的经典案例，帮助初学者理解 CNN 的基本原理和设计方法。


AlexNet通过暂退法控制全连接层的模型复杂度，而LeNet只使用了权重衰减。为了进一步扩充数
据，AlexNet在训练时增加了大量的图像增强数据，如翻转、裁切和变色。这使得模型更健壮，更大的样本量
有效地减少了过拟合。


```python
import torch
from torch import nn
from d2l import torch as d2l

net = nn.Sequential(
    # 这里使用一个11*11的更大窗口来捕捉对象。
    # 同时，步幅为4，以减少输出的高度和宽度。
    # 另外，输出通道的数目远大于LeNet
    nn.Conv2d(1, 96, kernel_size=11, stride=4, padding=1), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2),
    # 减小卷积窗口，使用填充为2来使得输入与输出的高和宽一致，且增大输出通道数
    nn.Conv2d(96, 256, kernel_size=5, padding=2), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2),
    # 使用三个连续的卷积层和较小的卷积窗口。
    # 除了最后的卷积层，输出通道的数量进一步增加。
    # 在前两个卷积层之后，汇聚层不用于减少输入的高度和宽度
    nn.Conv2d(256, 384, kernel_size=3, padding=1), nn.ReLU(),
    nn.Conv2d(384, 384, kernel_size=3, padding=1), nn.ReLU(),
    nn.Conv2d(384, 256, kernel_size=3, padding=1), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2),
    nn.Flatten(),
    # 这里，全连接层的输出数量是LeNet中的好几倍。使用dropout层来减轻过拟合
    nn.Linear(6400, 4096), nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(4096, 4096), nn.ReLU(),
    nn.Dropout(p=0.5),
    # 最后是输出层。由于这里使用Fashion-MNIST，所以用类别数为10，而非论文中的1000
    nn.Linear(4096, 10))
```

我们构造一个高度和宽度都为224的单通道数据，来观察每一层输出的形状。

```python
X = torch.randn(1, 1, 224, 224)
for layer in net:
    X=layer(X)
    print(layer.__class__.__name__,'output shape:\t',X.shape)
```

```python
batch_size = 128
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size, resize=224)
```

```python
lr, num_epochs = 0.01, 10
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

# 使用块的网络（VGG）

与芯片设计中工程师从放置晶体管到逻辑元件再到逻辑块的过程类似，神经网络架构的设计也逐渐变得更加
抽象。研究人员开始从单个神经元的角度思考问题，发展到整个层，现在又转向块，重复层的模式。

使用块的想法首先出现在牛津大学的视觉几何组（visual geometry group）91的VGG网络中。通过使用循环
和子程序，可以很容易地在任何现代深度学习框架的代码中实现这些重复的架构。

## VGG网络


### 1. VGG网络的背景
VGG网络是由牛津大学Visual Geometry Group（VGG）在2014年提出的，论文标题为《Very Deep Convolutional Networks for Large-Scale Image Recognition》。该网络在当年的ImageNet图像分类挑战赛（ILSVRC-2014）中取得了优异的成绩，尤其是在分类任务中表现突出。

VGG网络的核心思想是：**通过增加网络的深度（层数）来提高模型的性能**。为了验证这一点，VGG团队设计了多个不同深度的网络（如VGG11、VGG13、VGG16、VGG19等），并通过实验证明，**更深的网络可以提取更复杂的特征，从而提高分类精度**。

---

### 2. VGG网络的设计思想

#### 2.1 使用小卷积核
VGG网络的一个重要特点是**全部使用3x3的小卷积核**，而不是像之前的AlexNet那样使用较大的卷积核（如7x7或11x11）。这种设计有以下优点：
- **减少参数量**：3x3卷积核的参数数量较少，计算效率更高。
- **增加非线性**：通过堆叠多个3x3卷积层，可以在保持相同感受野的情况下，引入更多的非线性激活函数（ReLU），从而增强模型的表达能力。
- **捕捉局部特征**：小卷积核更适合捕捉图像的局部特征。

> **为什么3x3卷积核更好？**
> - 两个3x3卷积层的堆叠，其感受野相当于一个5x5卷积层。
> - 三个3x3卷积层的堆叠，其感受野相当于一个7x7卷积层。
> - 但参数量更少，且引入了更多的非线性。

#### 2.2 增加网络深度
VGG网络通过堆叠多个卷积层来增加网络的深度。常见的VGG16和VGG19分别有16层和19层（包括卷积层和全连接层）。实验表明，**更深的网络可以提取更高层次的特征**，从而提高模型的性能。

#### 2.3 池化层
VGG网络使用**2x2的最大池化层**，步幅为2。池化层的作用是逐步降低特征图的尺寸，同时保留重要的特征信息。

#### 2.4 全连接层
VGG网络的最后部分包含**3个全连接层**：
- 前两个全连接层各有4096个神经元。
- 最后一个全连接层有1000个神经元（对应ImageNet数据集的1000个类别）。
- 全连接层之后是Softmax层，用于输出分类概率。

---

### 3. VGG网络的具体结构

以VGG16为例，其详细结构如下：

| 层类型         | 配置                                | 输出尺寸（H x W x C） |

| 卷积层         | 256个3x3卷积核，步幅1，填充1        | 56x56x256             |
| 卷积层         | 256个3x3卷积核，步幅1，填充1        | 56x56x256             |

| 全连接层       | 1000个神经元                        | 1x1x1000              |
| Softmax层      | 输出分类概率                        | 1x1x1000              |

---

### 4. VGG网络的优点

1. **结构简单**：VGG网络的结构非常规整，易于理解和实现。
2. **性能优异**：在ImageNet等大规模数据集上表现优异，尤其是在分类任务中。
3. **适合迁移学习**：VGG网络的预训练模型可以用于其他计算机视觉任务（如目标检测、语义分割等），作为特征提取器。

---

### 5. VGG网络的缺点

1. **参数量大**：VGG网络的参数量非常大，尤其是全连接层占据了大部分参数。例如，VGG16的参数量约为1.38亿。
2. **计算量大**：由于网络较深，训练和推理需要大量的计算资源。
3. **内存占用高**：VGG网络的特征图尺寸较大，尤其是在浅层，导致内存占用较高。

---

### 6. VGG网络的意义

1. **证明了深度的重要性**：VGG网络通过实验证明了增加网络深度可以提高模型性能，为后续更深的网络（如ResNet）奠定了基础。
2. **成为经典基准模型**：VGG网络因其简单性和优异的性能，成为深度学习领域的经典基准模型之一。
3. **推动了迁移学习的发展**：VGG的预训练模型被广泛用于迁移学习，极大地推动了计算机视觉领域的发展。

现在来做相关实现

经典卷积神经网络的基本组成部分是下面的这个序列：

1. 带填充以保持分辨率的卷积层；
2. 非线性激活函数，如ReLU；
3. 汇聚层，如最大汇聚层

而一个VGG块与之类似，由一系列卷积层组成，后面再加上用于空间下采样的最大汇聚层。在最初的VGG论
文中 (Simonyan and Zisserman, 2014)，作者使用了带有3 × 3卷积核、填充为1（保持高度和宽度）的卷积层
和带有2 × 2汇聚窗口、步幅为2（每个块后的分辨率减半）的最大汇聚层。在下面的代码中，我们定义一
个名为vgg_block的函数来实现一个VG块。
该函数有三个参数，分别对应于卷积层的数量num_convs、输入通道的数量in_channels 和输出量的数
量out_channels.

```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
def vgg_block(num_convs, in_channels, out_channels):
    """
    定义一个VGG网络的卷积块。

    参数:
    - num_convs: 卷积层的数量。
    - in_channels: 输入特征图的通道数。
    - out_channels: 输出特征图的通道数。

    返回:
    - 一个包含多个卷积层、ReLU激活函数和最大池化层的Sequential模块。
    """
    layers = []  # 初始化一个空列表，用于存储每一层的操作

    # 循环创建卷积层和ReLU激活函数
    for _ in range(num_convs):
        # 添加一个卷积层
        layers.append(nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1))
        # 添加ReLU激活函数
        layers.append(nn.ReLU())
        # 更新输入通道数，以便下一层使用
        in_channels = out_channels

    # 添加最大池化层
    layers.append(nn.MaxPool2d(kernel_size=2, stride=2))

    # 将列表中的层按顺序组合成一个Sequential模块
    return nn.Sequential(*layers)
```

与AlexNet、LeNet一样，VGG网络可以分为两部分：第一部分主要由卷积层和汇聚层组成，第二部分由全连
接层组成。

```python
conv_arch = ((1, 64), (1, 128), (2, 256), (2, 512), (2, 512))
```

```python
def vgg(conv_arch):
    """
    定义一个完整的VGG网络，包括卷积部分和全连接部分。

    参数:
    - conv_arch: 一个元组，描述每个卷积块的配置，格式为 ((num_convs, out_channels), ...)。

    返回:
    - 一个包含卷积部分和全连接部分的Sequential模块。
    """
    conv_blks = []  # 初始化一个空列表，用于存储所有的卷积块
    in_channels = 1  # 输入图像的通道数（假设为1，例如灰度图像）

    # 卷积层部分
    for (num_convs, out_channels) in conv_arch:
        # 调用vgg_block函数创建卷积块，并将其添加到conv_blks列表中
        conv_blks.append(vgg_block(num_convs, in_channels, out_channels))
        # 更新输入通道数，作为下一个卷积块的输入
        in_channels = out_channels

    # 返回完整的VGG网络
    return nn.Sequential(
        *conv_blks,  # 展开conv_blks列表，将所有卷积块按顺序组合
        nn.Flatten(),  # 将多维的特征图展平为一维向量，以便输入全连接层

        # 全连接层部分
        nn.Linear(out_channels * 7 * 7, 4096),  # 第一个全连接层，输入大小为out_channels * 7 * 7，输出大小为4096
        nn.ReLU(),  # ReLU激活函数
        nn.Dropout(0.5),  # Dropout层，丢弃率为0.5，用于防止过拟合

        nn.Linear(4096, 4096),  # 第二个全连接层，输入大小为4096，输出大小为4096
        nn.ReLU(),  # ReLU激活函数
        nn.Dropout(0.5),  # Dropout层，丢弃率为0.5

        nn.Linear(4096, 10)  # 最后一个全连接层，输入大小为4096，输出大小为10（假设为10类分类任务）
    )

# 创建VGG网络实例
net = vgg(conv_arch)

# 打印网络结构
print(net)
```

接下来，我们将构建一个高度和宽度为224的单通道数据样本，以观察每个层输出的形状。

```python
X = torch.randn(size=(1, 1, 224, 224))
for blk in net:
    X = blk(X)
    print(blk.__class__.__name__,'output shape:\t',X.shape)
```

正如从代码中所看到的，我们在每个块的高度和宽度减半，最终高度和宽度都为7。最后再展平表示，送入全连接层处理。


由于VGG‐11比AlexNet计算量更大，因此我们构建了一个通道数较少的网络，足够用于训Fashion‐MNIST数据集。

```python
# 定义一个比例因子，用于缩小VGG网络中每个卷积块的输出通道数
ratio = 4

# 使用列表推导式，根据比例因子缩小conv_arch中每个卷积块的输出通道数
small_conv_arch = [(pair[0], pair[1] // ratio) for pair in conv_arch]

# 使用缩小后的卷积块配置创建VGG网络实例
net = vgg(small_conv_arch)
```

除了使用略高的学习率外，模型训练过程与AlexNet类似。

```python
lr, num_epochs, batch_size = 0.05, 10, 128
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size, resize=224)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

# 网络中的网络（NiN）


## 1. NiN 的设计动机

在传统的卷积神经网络（CNN）中，卷积层通过线性滤波器（卷积核）提取局部特征，然后通过非线性激活函数（如ReLU）引入非线性。然而，这种线性卷积操作的特征提取能力有限，尤其是当数据分布复杂时，简单的线性变换可能无法充分捕捉数据的特征。

**NiN的核心思想**是：在每个局部感受野上，使用一个**多层感知机（MLP）**来替代传统的线性卷积操作。MLP本身是一个非线性函数，能够更好地拟合复杂的特征分布。通过这种方式，NiN增强了模型的非线性表达能力。

---

## 2. NiN 的具体结构

NiN的结构主要由以下几个关键组件构成：

### （1）MLP 卷积层
- 传统的卷积层使用线性滤波器（卷积核）对输入进行线性变换，然后通过激活函数引入非线性。
- NiN的MLP卷积层则使用一个小型的MLP（通常是一个两层的全连接网络）来替代线性滤波器。这个MLP的输入是局部感受野的像素值，输出是特征图的一个点。
- MLP的结构通常为：**全连接层 + ReLU激活函数 + 全连接层 + ReLU激活函数**。
- 通过这种方式，MLP卷积层能够捕捉更复杂的局部特征。

### （2）1x1 卷积
- NiN广泛使用了**1x1卷积**。1x1卷积的作用主要有两个：
  1. **降维**：通过减少通道数，降低计算复杂度。
  2. **增加非线性**：1x1卷积可以看作是对通道的线性组合，再通过激活函数引入非线性。
- 1x1卷积在NiN中起到了至关重要的作用，它不仅减少了参数数量，还增强了模型的表达能力。

### （3）全局平均池化（Global Average Pooling, GAP）
- 传统的CNN通常在最后使用全连接层进行分类，但全连接层会引入大量参数，容易导致过拟合。
- NiN在最后一层使用**全局平均池化**，即对每个特征图取平均值，得到一个标量。这些标量直接作为分类器的输入。
- 全局平均池化的优点：
  1. **减少参数**：避免了全连接层的大量参数。
  2. **防止过拟合**：全局平均池化是一种正则化手段，能够有效防止过拟合。
  3. **空间鲁棒性**：全局平均池化对输入图像的空间位置不敏感，增强了模型的鲁棒性。

### （4）NiN 的总体架构
- NiN的典型结构由多个**NiN块**堆叠而成，每个NiN块包含：
  1. 一个MLP卷积层。
  2. 一个1x1卷积层。
  3. 一个最大池化层（用于下采样）。
- 最后，NiN使用全局平均池化层替代全连接层，直接输出分类结果。

---

## 3. NiN 的数学原理

### （1）MLP 卷积层的数学表达
假设输入特征图为 $X \in \mathbb{R}^{H \times W \times C}$，其中 $H$ 和 $W$ 是空间维度，$C$ 是通道数。MLP卷积层的操作可以表示为：
$$
Y_{i,j,k} = f\left(\sum_{c=1}^C W_{k,c} \cdot X_{i,j,c} + b_k\right)
$$
其中：
- $W_{k,c}$ 是MLP的权重。
- $b_k$ 是偏置。
- $f$ 是非线性激活函数（如ReLU）。

### （2）1x1 卷积的数学表达
1x1卷积的操作可以表示为：
$$
Y_{i,j,k} = \sum_{c=1}^C W_{k,c} \cdot X_{i,j,c} + b_k
$$
其中：
- $W_{k,c}$ 是1x1卷积的权重。
- $b_k$ 是偏置。

### （3）全局平均池化的数学表达
全局平均池化的操作可以表示为：
$$
y_k = \frac{1}{H \times W} \sum_{i=1}^H \sum_{j=1}^W X_{i,j,k}
$$
其中：
- $y_k$ 是第 $k$ 个特征图的平均值。
- $H$ 和 $W$ 是特征图的空间维度。

---

## 4. NiN 的优缺点

### 优点
1. **增强非线性**：MLP卷积层能够捕捉更复杂的特征。
2. **减少参数**：1x1卷积和全局平均池化显著减少了参数量。
3. **防止过拟合**：全局平均池化是一种有效的正则化手段。
4. **空间鲁棒性**：全局平均池化对输入图像的空间位置不敏感。

### 缺点
1. **计算复杂度高**：MLP卷积层增加了计算负担。
2. **实现复杂**：相比传统卷积，NiN的实现更为复杂。

---

## 5. NiN 的实际应用

NiN在多个视觉任务中取得了良好的效果，尤其是在小规模数据集（如CIFAR-10和CIFAR-100）上表现优异。此外，NiN的设计思想（如1x1卷积和全局平均池化）被后续的模型（如GoogLeNet和ResNet）广泛采用。


```python
import torch
from torch import nn
from d2l import torch as d2l

# 定义一个NiN块（NiN Block）
def nin_block(in_channels, out_channels, kernel_size, strides, padding):
    """
    定义一个NiN块，包含以下层：
    1. 一个卷积层（用于提取特征）。
    2. 两个1x1卷积层（用于增强非线性表达能力）。
    每个卷积层后都接一个ReLU激活函数。

    参数:
    - in_channels: 输入特征图的通道数。
    - out_channels: 输出特征图的通道数。
    - kernel_size: 卷积核的大小。
    - strides: 卷积的步幅。
    - padding: 卷积的填充大小。

    返回:
    - 一个nn.Sequential模块，包含定义的NiN块。
    """
    return nn.Sequential(
        # 第一层：常规卷积层
        nn.Conv2d(in_channels, out_channels, kernel_size, strides, padding),
        nn.ReLU(),  # ReLU激活函数
        
        # 第二层：1x1卷积层
        nn.Conv2d(out_channels, out_channels, kernel_size=1),  # 1x1卷积，用于降维和增强非线性
        nn.ReLU(),  # ReLU激活函数
        
        # 第三层：1x1卷积层
        nn.Conv2d(out_channels, out_channels, kernel_size=1),  # 1x1卷积，进一步增强非线性
        nn.ReLU()  # ReLU激活函数
    )
```

NiN和AlexNet之间的一个显著区别是NiN完全取消了全连接层。相反，NiN使用一个NiN块，其输出通道数等
于标签类别的数量。最后放一个全局平均汇聚层（global average pooling layer），生成一个对数几率（logits）。
NiN设计的一个优点是，它显著减少了模型所需参数的数量。然而，在实践中，这种设计有时会增加训练模
型的时间。


# 定义一个完整的NiN网络
net = nn.Sequential(
    # 第一个NiN块
    nin_block(1, 96, kernel_size=11, strides=4, padding=0),  # 输入通道数为1，输出通道数为96，卷积核大小为11x11，步幅为4，无填充
    nn.MaxPool2d(3, stride=2),  # 最大池化层，池化核大小为3x3，步幅为2，用于下采样
    
    # 第二个NiN块
    nin_block(96, 256, kernel_size=5, strides=1, padding=2),  # 输入通道数为96，输出通道数为256，卷积核大小为5x5，步幅为1，填充为2
    nn.MaxPool2d(3, stride=2),  # 最大池化层，池化核大小为3x3，步幅为2，用于下采样
    
    # 第三个NiN块
    nin_block(256, 384, kernel_size=3, strides=1, padding=1),  # 输入通道数为256，输出通道数为384，卷积核大小为3x3，步幅为1，填充为1
    nn.MaxPool2d(3, stride=2),  # 最大池化层，池化核大小为3x3，步幅为2，用于下采样
    
    # Dropout层
    nn.Dropout(0.5),  # Dropout层，丢弃率为0.5，用于防止过拟合
    
    # 第四个NiN块
    nin_block(384, 10, kernel_size=3, strides=1, padding=1),  # 输入通道数为384，输出通道数为10，卷积核大小为3x3，步幅为1，填充为1
    # 注意：这里的输出通道数为10，对应分类任务的类别数
    
    # 全局自适应平均池化层
    nn.AdaptiveAvgPool2d((1, 1)),  # 全局自适应平均池化层，将每个特征图的空间维度压缩为1x1
    
    # 展平层
    nn.Flatten()  # 将四维的输出（批量大小, 通道数, 1, 1）转成二维的输出（批量大小, 通道数）
)

创建一个数据样本来查看每个块的输出形状。

```python
X = torch.rand(size=(1, 1, 224, 224))
for layer in net:
    X = layer(X)
    print(layer.__class__.__name__,'output shape:\t', X.shape)
```

使用Fashion‐MNIST来训练模型。训练NiN与训练AlexNet、VGG时相似。

```python
lr, num_epochs, batch_size = 0.1, 10, 128
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size, resize=224)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

# 含并行连结的网络（GoogLeNet）

## 1. **GoogLeNet 的设计动机**

GoogLeNet 是由 Google 团队在 2014 年提出的深度卷积神经网络，其主要目标是：
1. **提高模型的表达能力**：通过并行结构捕捉多尺度的特征。
2. **减少参数数量**：通过 1x1 卷积降维，减少计算复杂度。
3. **适应不同的输入尺度**：通过多尺度卷积核提取特征。

GoogLeNet 的核心创新是 **Inception 模块**，它通过并行使用不同大小的卷积核和池化操作，增强了模型的多尺度特征提取能力。

---

## 2. **Inception 模块**

Inception 模块是 GoogLeNet 的核心组件，其设计思想是**在同一层级上并行使用不同大小的卷积核和池化操作**，从而捕捉多尺度的特征。

### Inception 模块的结构
一个典型的 Inception 模块包含以下并行分支：
1. **1x1 卷积**：用于降维和增强非线性。
2. **3x3 卷积**：捕捉中等尺度的特征。
3. **5x5 卷积**：捕捉更大尺度的特征。
4. **3x3 最大池化**：用于下采样，保留重要特征。

每个分支的输出在通道维度上拼接（concat），形成最终的特征图。

### Inception 模块的改进
为了减少计算量，GoogLeNet 在 Inception 模块中引入了 **1x1 卷积**：
- 在 3x3 和 5x5 卷积之前，使用 1x1 卷积降维，减少通道数。
- 在池化分支之后，使用 1x1 卷积调整通道数。

这种设计不仅减少了参数量，还增强了模型的非线性表达能力。

---

## 3. **GoogLeNet 的整体结构**

GoogLeNet 的整体结构由多个 Inception 模块堆叠而成，中间穿插了池化层用于下采样。其结构如下：

### （1）初始卷积层
- 使用一个 7x7 卷积层提取初步特征，后接一个最大池化层进行下采样。

### （2）Inception 模块堆叠
- 多个 Inception 模块用于提取多尺度特征。
- 每个 Inception 模块的输入和输出通道数逐渐增加，以捕捉更复杂的特征。

### （3）辅助分类器
- 在网络中间插入两个辅助分类器，用于缓解梯度消失问题。
- 辅助分类器在训练时提供额外的梯度信号，帮助深层网络的训练。

### （4）全局平均池化层
- 在网络的最后，使用全局平均池化层替代全连接层。
- 全局平均池化层将每个特征图的空间维度压缩为 1x1，输出形状为 `(批量大小, 通道数, 1, 1)`。

### （5）全连接层
- 最后通过一个全连接层输出分类结果。

---

## 4. **GoogLeNet 的优缺点**

### 优点
1. **多尺度特征提取**：Inception 模块通过并行结构捕捉多尺度特征。
2. **参数效率高**：通过 1x1 卷积降维，减少了参数量和计算量。
3. **防止过拟合**：辅助分类器和全局平均池化层有效防止了过拟合。
4. **适应不同输入尺寸**：全局平均池化层可以接受任意大小的输入特征图。

#### 缺点
1. **结构复杂**：Inception 模块的设计和实现较为复杂。
2. **训练难度大**：由于网络较深，训练时需要更多的技巧（如辅助分类器）。

---

## 5. **GoogLeNet 的应用**

GoogLeNet 在 ImageNet 分类任务中取得了优异的成绩，其设计思想（如 Inception 模块和 1x1 卷积）被广泛应用于后续的深度学习模型中（如 Inception-v2、Inception-v3 等）。

---

## 6. **总结**

GoogLeNet 通过引入 Inception 模块，实现了多尺度特征提取和参数高效利用。其核心思想包括：
1. **并行结构**：通过不同大小的卷积核和池化操作捕捉多尺度特征。
2. **1x1 卷积**：用于降维和增强非线性。
3. **全局平均池化**：替代全连接层，减少参数量。
4. **辅助分类器**：缓解梯度消失问题，帮助深层网络的训练。

```python
import torch
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l
```

```python
class Inception(nn.Module):
    # c1--c4是每条路径的输出通道数
    def __init__(self, in_channels, c1, c2, c3, c4, **kwargs):
        """
        Inception 模块的初始化函数。

        参数:
        - in_channels: 输入特征图的通道数。
        - c1: 线路1的输出通道数（1x1卷积）。
        - c2: 线路2的输出通道数（1x1卷积 + 3x3卷积），c2[0]是1x1卷积的输出通道数，c2[1]是3x3卷积的输出通道数。
        - c3: 线路3的输出通道数（1x1卷积 + 5x5卷积），c3[0]是1x1卷积的输出通道数，c3[1]是5x5卷积的输出通道数。
        - c4: 线路4的输出通道数（3x3最大池化 + 1x1卷积）。
        """
        super(Inception, self).__init__(**kwargs)
        
        # 线路1，单1x1卷积层
        self.p1_1 = nn.Conv2d(in_channels, c1, kernel_size=1)
        
        # 线路2，1x1卷积层后接3x3卷积层
        self.p2_1 = nn.Conv2d(in_channels, c2[0], kernel_size=1)  # 1x1卷积降维
        self.p2_2 = nn.Conv2d(c2[0], c2[1], kernel_size=3, padding=1)  # 3x3卷积
        
        # 线路3，1x1卷积层后接5x5卷积层
        self.p3_1 = nn.Conv2d(in_channels, c3[0], kernel_size=1)  # 1x1卷积降维
        self.p3_2 = nn.Conv2d(c3[0], c3[1], kernel_size=5, padding=2)  # 5x5卷积
        
        # 线路4，3x3最大池化层后接1x1卷积层
        self.p4_1 = nn.MaxPool2d(kernel_size=3, stride=1, padding=1)  # 3x3最大池化
        self.p4_2 = nn.Conv2d(in_channels, c4, kernel_size=1)  # 1x1卷积调整通道数

    def forward(self, x):
        """
        Inception 模块的前向传播函数。

        参数:
        - x: 输入特征图，形状为 (批量大小, 输入通道数, 高度, 宽度)。

        返回:
        - 输出特征图，形状为 (批量大小, c1 + c2[1] + c3[1] + c4, 高度, 宽度)。
        """
        # 线路1：单1x1卷积层
        p1 = F.relu(self.p1_1(x))  # 1x1卷积 + ReLU激活
        
        # 线路2：1x1卷积层后接3x3卷积层
        p2 = F.relu(self.p2_2(F.relu(self.p2_1(x))))  # 1x1卷积 + ReLU激活 + 3x3卷积 + ReLU激活
        
        # 线路3：1x1卷积层后接5x5卷积层
        p3 = F.relu(self.p3_2(F.relu(self.p3_1(x))))  # 1x1卷积 + ReLU激活 + 5x5卷积 + ReLU激活
        
        # 线路4：3x3最大池化层后接1x1卷积层
        p4 = F.relu(self.p4_2(self.p4_1(x)))  # 3x3最大池化 + 1x1卷积 + ReLU激活
        
        # 在通道维度上拼接输出
        return torch.cat((p1, p2, p3, p4), dim=1)
```

GoogLeNet一共使用9个Inception块和全局平均汇聚层的堆叠来生成其估计值。Inception块
之间的最大汇聚层可降低维度。第一个模块类似于AlexNet和LeNet，Inception块的组合从VGG继承，全局平
均汇聚层避免了在最后使用全连接层。

逐一实现GoogLeNet的每个模块。

第一个模块使用64个通道、7 × 7卷积层。

```python
b1 = nn.Sequential(nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3),
                    nn.ReLU(),
                    nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
```

第二个模块使用两个卷积层：第一个卷积层是64个通道、1 × 1卷积层；第二个卷积层使用将通道数量增加三
倍的3 × 3卷积层。这对应于Inception块中的第二条路径。

```python
b2 = nn.Sequential(nn.Conv2d(64, 64, kernel_size=1),
                    nn.ReLU(),
                    nn.Conv2d(64, 192, kernel_size=3, padding=1),
                    nn.ReLU(),
                    nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
```

第三个模块串联两个完整的Inception块。第一个Inception块的输出通道数为64 + 128 + 32 + 32 = 256，四
个路径之间的输出通道数量比为64 : 128 : 32 : 32 = 2 : 4 : 1 : 1。第二个和第三个路径首先将输入通道的数量
分别减少到96/192 = 1/2和16/192 = 1/12，然后连接第二个卷积层。第二个Inception块的输出通道数增加
到128 + 192 + 96 + 64 = 480，四个路径之间的输出通道数量比为128 : 192 : 96 : 64 = 4 : 6 : 3 : 2。第二条和
第三条路径首先将输入通道的数量分别减少到128/256 = 1/2和32/256 = 1/8。

```python
b3 = nn.Sequential(Inception(192, 64, (96, 128), (16, 32), 32),
                    Inception(256, 128, (128, 192), (32, 96), 64),
                    nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
```

第四模块更加复杂，它串联了5个Inception块，其输出通道数分别是192 + 208 + 48 + 64 = 512、160 + 224 +
64 + 64 = 512、128 + 256 + 64 + 64 = 512、112 + 288 + 64 + 64 = 528和256 + 320 + 128 + 128 = 832。这些
路径的通道数分配和第三模块中的类似，首先是含3×3卷积层的第二条路径输出最多通道，其次是仅含1×1卷
积层的第一条路径，之后是含5×5卷积层的第三条路径和含3×3最大汇聚层的第四条路径。其中第二、第三条
路径都会先按比例减小通道数。这些比例在各个Inception块中都略有不同。

```python
b4 = nn.Sequential(Inception(480, 192, (96, 208), (16, 48), 64),
                    Inception(512, 160, (112, 224), (24, 64), 64),
                    Inception(512, 128, (128, 256), (24, 64), 64),
                    Inception(512, 112, (144, 288), (32, 64), 64),
                    Inception(528, 256, (160, 320), (32, 128), 128),
                    nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
```

第五模块包含输出通道数为256 + 320 + 128 + 128 = 832和384 + 384 + 128 + 128 = 1024的两个Inception块。
其中每条路径通道数的分配思路和第三、第四模块中的一致，只是在具体数值上有所不同。需要注意的是，第
五模块的后面紧跟输出层，该模块同NiN一样使用全局平均汇聚层，将每个通道的高和宽变成1。最后我们将
输出变成二维数组，再接上一个输出个数为标签类别数的全连接层。

```python
b5 = nn.Sequential(Inception(832, 256, (160, 320), (32, 128), 128),
                    Inception(832, 384, (192, 384), (48, 128), 128),
                    nn.AdaptiveAvgPool2d((1,1)),
                    nn.Flatten())

net = nn.Sequential(b1, b2, b3, b4, b5, nn.Linear(1024, 10))
```

```python
X = torch.rand(size=(1, 1, 96, 96))

for layer in net:
    X = layer(X)
    print(layer.__class__.__name__,'output shape:\t', X.shape)
```
